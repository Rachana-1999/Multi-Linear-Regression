# Preparing a model

```python
import statsmodels.formula.api as smf 
model = smf.ols('Price~Age_08_04+KM+HP+cc+Doors+Gears+Quarterly_Tax+Weight',data=toy).fit()
```

```python
model.summary()
```

```python
print(model.pvalues)
```

# Simple Linear Regression Models

```python
ml_v=smf.ols('Price~cc',data = toy).fit()  
print( ml_v.pvalues)  
```

```python
ml_v=smf.ols('Price~Doors',data = toy).fit()  
print( ml_v.pvalues)  
```

```python
ml_v=smf.ols('Price~cc+Doors',data = toy).fit()  
print(ml_v.pvalues)  
```

# Q-Q Plot

```python
import statsmodels.api as sm
qqplot=sm.qqplot(model.resid,line='q') 
plt.title("Normal Q-Q plot of residuals")
plt.show()
```

```python
list(np.where(model.resid>500))
```

# Residual Plot for Homoscedasticity


```python
def get_standardized_values( vals ):
    return (vals - vals.mean())/vals.std()
```

```python
plt.scatter(get_standardized_values(model.fittedvalues),
            get_standardized_values(model.resid))

plt.title('Residual Plot')
plt.xlabel('Standardized Fitted values')
plt.ylabel('Standardized residual values')
plt.show()
```

#  Residual Vs Regressors

```python
fig = plt.figure(figsize=(15,8))
fig = sm.graphics.plot_regress_exog(model, "cc", fig=fig)
plt.show()
```

```python
fig = plt.figure(figsize=(15,8))
fig = sm.graphics.plot_regress_exog(model, "Gears", fig=fig)
plt.show()
```

# Cooks Distance 

```python
model_influence = model.get_influence()
(c, _) = model_influence.cooks_distance
```

```python
fig = plt.subplots(figsize=(20, 7))
plt.stem(np.arange(len(toy)), np.round(c, 3))
plt.xlabel('Row index')
plt.ylabel('Cooks Distance')
plt.show()
```

```python
(np.argmax(c),np.max(c))
```
